```python
import pandas as pd
import requests
import zipfile
import io
import sqlite3
#from sqlite3 import Error
import datetime
```

```python
link2FFcsvZip = 'http://fracfocusdata.org/digitaldownload/FracFocusCSV.zip'
```

```python
request = requests.get(link2FFcsvZip)
zipBag = zipfile.ZipFile(io.BytesIO(request.content))
```

```python
zipFiles = zipBag.namelist()
zipFiles
```

```python
FracFocusRegistry_files=[]
RegistryUpload_files=[]
```

```python
for file_name in zipFiles:
    if (('FracFocusRegistry' in file_name)) & (file_name.endswith('.csv')):
        FracFocusRegistry_files.append(file_name)
    if (('registryupload' in file_name)) & (file_name.endswith('.csv')):
        RegistryUpload_files.append(file_name)
```

```python
FracFocusRegistry_files
```

```python
RegistryUpload_files
```

```python
dfs = list()
for file_name in FracFocusRegistry_files:
    df = pd.read_csv(zipBag.open(file_name), low_memory=False, dtype={'APINumber': 'str','StateNumber': 'str', 'CountyNumber': 'str'})
    dfs.append(df)

FracFocusRegistry = pd.concat(dfs, axis=0, ignore_index=True)
FracFocusRegistry.head(2)
```

```python
dfs = list()
for file_name in RegistryUpload_files:
    df = pd.read_csv(zipBag.open(file_name), low_memory=False, dtype={'APINumber': 'str','StateNumber': 'str', 'CountyNumber': 'str'})
    dfs.append(df)

RegistryUpload = pd.concat(dfs, axis=0, ignore_index=True)
RegistryUpload.head(2)
```

```python
ffWellsND = RegistryUpload[['pKey','APINumber','JobEndDate','TotalBaseWaterVolume','TotalBaseNonWaterVolume']][RegistryUpload['StateNumber']== '33'].reset_index(drop=True)
```

```python
ffWellsND['JobEndDate'] = pd.to_datetime(ffWellsND['JobEndDate'], errors = 'coerce').dt.date
```

```python
ffWellsND.tail()
```

```python
swData = FracFocusRegistry[['UploadKey','APINumber','JobEndDate','IngredientName','Purpose','TradeName','MassIngredient','PercentHFJob','IngredientKey']][FracFocusRegistry['StateNumber']== '33'].reset_index(drop=True)
swData['JobEndDate'] = pd.to_datetime(swData['JobEndDate'], errors = 'coerce').dt.date
```

```python
swData.tail()
```

```python
exclusionList = ['water','chemical', 'acid','stabilizer','carrier','clay','ammonium','chloride','Potassium','KCL','chlorine','naphtha','petroleum','aromatic','surfactant','alcohol','gum','tetrahydrate','carbonate','glycol','Other','salt','Inhibitor','Additives','Additive', 'Sulfate','Gel','Biocides','pH','Proprietary','Biocide','Viscosifier','Activator','Solvent','Methanol','Conductivity']
```

```python
swDataFilter = swData[~swData['IngredientName'].str.contains('|'.join(exclusionList), case=False, na=False)].copy()
```

```python
swDataFilter = swDataFilter[~swDataFilter['Purpose'].str.contains('|'.join(exclusionList), case=False, na=False)].copy()
```

```python
swDataFilter = swDataFilter[~swDataFilter['TradeName'].str.contains('|'.join(exclusionList), case=False, na=False)].copy()
```

```python
swDataFilter = swDataFilter.dropna().reset_index(drop = True)
```

```python
swDataFilter.tail(60)
```
