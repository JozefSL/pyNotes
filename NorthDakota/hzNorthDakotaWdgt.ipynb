```python
filename= 'OGD_Horizontals.dbf'

```

```python
# write to csv file about 2 minutes
table = DBF(filename)
f=open('ndWells.csv', 'w')
writer = csv.writer(f)
writer.writerow(table.field_names)
for record in table:
  writer.writerow(list(record.values()))
```
